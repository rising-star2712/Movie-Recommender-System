```python
# multiple dictionaries to list of names of genres , ex. [Action , Adventure ,...]
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj) :
    L.append(i['name'])
  return L
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj) :
   if counter != 3:
      L.append(i['name'])
      counter+=1
   else :
     break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj) :
   if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
  
```

```python
 movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " , "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " , "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " , "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id' , 'title' , 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_params()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text) :
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)  
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
